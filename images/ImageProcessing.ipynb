## Resize Now
    


```python
from PIL import Image
for i in images:
    im = Image.open(i)
    w = 1920 if im.width > 1920 else im.width
    h = im.height * w // im.width 
    if w < im.width:
        print("Resizing {} to {} * {}".format(i, w, h))
        re = im.resize((w, h)) 
        re.save(i)
    else:
        print("Image {} is not big enough to convert..")
```
